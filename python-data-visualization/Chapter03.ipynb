```python
plt.hist(df['comb08']);
plt.xlabel('Combined Fuel Economy')
plt.ylabel('Number of cars')
plt.title('Combined Fuel Economy (2000-2020)')
plt.show()
```

### Using the object oriented API for matplotlib

```python
fig, ax = plt.subplots()
ax.hist(df['comb08']);
```

```python
fig, ax = plt.subplots()
ax.hist(df['comb08']);
ax.set_xlabel('Combined Fuel Economy (MPG)')
ax.set_ylabel('Number of cars')
ax.set_title('2000-2020')
```

```python
fig, ax = plt.subplots()
ax.hist(df['comb08']);
ax.set(xlabel='Combined Fuel Economy (MPG)', ylabel='Number of cars', title='2000-2020')
```

### Customizing the plot

```python
fig, ax = plt.subplots()
ax.hist(df['comb08'], range=(10,50));
```

```python
fig, ax = plt.subplots()
ax.hist(df['comb08'], range=(10,50), cumulative=True);
```

```python
fig, ax = plt.subplots()
ax.hist(df['comb08'], range=(10,50), histtype='step', orientation='horizontal');
```

```python
fig, ax = plt.subplots()
ax.hist(df['comb08'], range=(10,50), bins=10);
```

It is possible to get the values such as what the bins values are

```python
fig, ax = plt.subplots()
n, bins, patches = ax.hist(df['comb08'], range=(10,50), bins=10)
```
